```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA,KernelPCA
```

```python
data = pd.read_csv('../data/hmeq_clean_cat.csv',index_col=0)
```

```python
data.head()
```

```python
data_pos = data[data['BAD']==0]
data_neg = data[data['BAD']==1]
```

```python
%matplotlib inline
```

```python
pca = PCA(n_components=2)
kpca = KernelPCA(n_components=2,kernel='poly')
```

```python
pca.fit(data)
kpca.fit(data)
```

```python
pca_pos=pca.transform(data_pos)
kpca_pos=kpca.transform(data_pos)
```

```python
pca_neg=pca.transform(data_neg)
kpca_neg=kpca.transform(data_neg)
```

```python

plt.scatter(kpca_pos[:,0],kpca_pos[:,1],color="blue")
plt.scatter(kpca_neg[:,0],kpca_neg[:,1],color="red")
```
